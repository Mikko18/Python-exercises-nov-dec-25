```python
name = "Kyle"
guess_count = 0
while True:
    guess = input("Please, guess my name.")
    guess_count += 1
    if guess == name:
        print("Congratulations!")
        print(f"Guesses: {guess_count}")
        break
    else:
        quit_choice = input("Do you want to quit (y/n) ?")
        if quit_choice == "y":
            break
```
